```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

selected_columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'sqft_basement', 'sqft_above', 'floors', 'view', 'condition']
x = df[selected_columns]
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

lr = LinearRegression()

lr.fit(x_train, y_train)

predictions = lr.predict(x_test)

mse = mean_squared_error(y_test, predictions)
print('Mean Squared Error:', mse)
plt.figure(figsize=(8, 6))
plt.scatter(y_test, predictions)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predictions)
print('R-squared Score:', r2)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
features = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'city_encoded']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
```

# K means

```python
distortions = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    distortions.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.plot(K_range, distortions, marker='o')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion')
plt.show()
```

```python
for i in distortions:
  print(i)
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(principal_df['PC1'], principal_df['PC2'], c=df['cluster'], cmap='viridis', edgecolors='k', s=50, alpha=0.7)
plt.title('K-Means Clustering (2D Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
cluster_names = {0: 'Sunset Haven', 1: 'Cosmo Lofts', 2: 'Tranquil Grove'}
df['cluster_name'] = df['cluster'].map(cluster_names)
df = df.drop(columns =["cluster"], axis = 1)
```
